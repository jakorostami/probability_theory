```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import random
random.seed(1990)
```

```python
n = 10000
```

```python
X = np.random.uniform(size=n) # Generate 1000 values randomly from a uniform distribution
```

```python
sns.histplot(X) # Histogram plot confirming we stretch from 0 to 1 according to a U~(0, 1)
```

```python
U = -np.log(X)
```

```python
sns.histplot(U)
```

```python
# Compare it to a "regular" random Exp~(1)
Y = np.random.exponential(size=n)

sns.histplot(Y)
```

```python
fig, axes = plt.subplots(1, 2, sharex=True,
                         figsize=(8, 5))

fig.suptitle("Generating Exp~(1)")
axes[0].set_title("Exp from a Uniform (inverse transform)")
axes[1].set_title("Exp from np.random.exponential()")
sns.histplot(U, ax=axes[0])
sns.histplot(Y, ax=axes[1])
```
