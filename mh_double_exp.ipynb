```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Simulate from a double exponential with below density

```python
n = 10000
lmd = 4

simulations = []
```

```python
for i in range(n):
    state = 0
    
    for j in range(50):
        y = np.random.normal(size=1, loc=state, scale=1.0) # Using Gaussian as proposal distribution
        a = np.exp(-lmd * (abs(y) - abs(state) ) )
        
        if np.random.uniform() < a:
            state = y
    simulations.append(state)
```

```python
simulations = np.reshape(simulations, newshape=(10000,))
```

```python
sns.distplot(simulations, hist=True, kde=True)
```

```python
x = np.linspace(-np.pi, np.pi, 10000)

def double_exp(x):
    return lmd/2 * np.exp(-lmd * abs(x))
```

```python
plt.hist(simulations, density=True, bins=30)
plt.plot(x, double_exp(x))
plt.title("Simulating from the double exponential with MH algorithm")
```
