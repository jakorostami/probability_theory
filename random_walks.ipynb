```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
n=1000

def random_walk(n):
    return np.cumsum(np.random.normal(size=n))
```

```python
# 1D random walk

x = random_walk(n)
```

```python
plt.plot(x)
```

```python
# 2D random walk

x = random_walk(n)
y = random_walk(n)

plt.plot(x, y)
```

```python
# 3D random walk
n=2000

x = random_walk(n)
y = random_walk(n)
z = random_walk(n)

fig = plt.figure(figsize=(15,10))
axes = plt.subplot(1,1,1, projection="3d")
axes.plot(x, y , z)

axes.scatter(x[-1], y[-1], z[-1])


plt.show()
```
